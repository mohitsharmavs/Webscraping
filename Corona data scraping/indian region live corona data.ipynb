```python
def state_corona_update():
    import pandas as pd
    import numpy as np
    import requests
    url = "https://api.apify.com/v2/datasets/58a4VXwBBF0HtxuQa/items?format=json&clean=1"
    resp = requests.get(url)
    date = []
    data = resp.json()
    region=[]
    totalInfected=[]
    newInfected=[]
    recovered=[]
    newRecovered=[]
    deceased=[]
    newDeceased=[]
    for val in data[63:]:
        for i in val.get('regionData'):
            date.append(val.get('lastUpdatedAtApify',np.nan))
            region.append(i.get('region',np.nan))
            totalInfected.append(i.get('totalInfected',np.nan))
            newInfected.append(i.get('newInfected',np.nan))
            recovered.append(i.get('recovered',np.nan))
            newRecovered.append(i.get('newRecovered',np.nan))
            deceased.append(i.get('deceased',np.nan))
            newDeceased.append(i.get('newDeceased',np.nan))
#     global df       
    df = pd.DataFrame({
                "Date" : date,
                "Region" : region,
                "TotalInfected" : totalInfected,
                "NewInfected" : newInfected,
                "Recovered" : recovered,
                "NewRecovered" : newRecovered,
                "Deceased": deceased,
                "NewDeceased": newDeceased

                })
    df['Date'] = df[['Date']].applymap(lambda x: x[0:10])
    df['Date']=pd.to_datetime(df['Date'])

    return df
```

```python
df = state_corona_update()
```
